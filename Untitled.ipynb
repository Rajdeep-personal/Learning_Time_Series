```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Importing Data

```python
data = pd.read_csv('UNRATE.csv', header = None)
data.columns = ['Date','Number']
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data = data.set_index('Date')
data.head()
```

```python
data.head(15)
```

## Time Series Data

### Plotting Time Series Data

```python
data.plot(figsize=(12, 4))
plt.legend(loc='best')
plt.title('Unemployment Rate')
plt.show(block=False)
```

## Time Series Decomposition

### Additive Decomposition Method

```python
# from pylab import rcParams
plt.figure(figsize=(12,8))
import statsmodels.api as sm
#rcParams['figure.figsize'] = 12, 8
decomposition = sm.tsa.seasonal_decompose(data.Number, model='additive') # additive seasonal index
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

### Multiplicative Decomposition Method

```python
plt.figure(figsize=(12,8))
import statsmodels.api as sm
decomposition = sm.tsa.seasonal_decompose(data.Number, model='multiplicative') # multiplicative seasonal index
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

# Build and evaluate time series forecast

## Split time series data into training and test set

```python
data.shape
```

```python
train_len = 612
train = data[0:train_len] # first 612 months as training set
test = data[train_len:] # last 255 months as out-of-time test set
```

```python
train
```

```python
test
```

## Naive Method

```python
train.iloc[train_len-1]
```

```python
train['Number'][train_len-1]
```

```python
y_hat_naive = test.copy()
y_hat_naive['naive_forecast'] = train['Number'][train_len-1]
```

### Plot Train, Test and Forecast

```python
plt.figure(figsize=(12,4))
plt.plot(train['Number'], label='Train')
plt.plot(test['Number'], label='Test')
plt.plot(y_hat_naive['naive_forecast'], label='Naive forecast')
plt.legend(loc='best')
plt.title('Naive Method')
plt.show()
```

### Calculate RMSE and MAPE

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(test['Number'], y_hat_naive['naive_forecast'])).round(2)
mape = np.round(np.mean(np.abs(test['Number']-y_hat_naive['naive_forecast'])/test['Number'])*100,2)

results = pd.DataFrame({'Method':['Naive method'], 'MAPE': [mape], 'RMSE': [rmse]})
results = results[['Method', 'RMSE', 'MAPE']]
results
```

### Stationarity Check

### ADF(Augmented Dickey Fuller Test)

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(data['Number'])
print(adf_test)
```

```python
print('ADF statistic: %f' % adf_test[0])
print('Critical Value @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %.2f' % adf_test[1])
```

- p-value is less than 0.05 , which shows the series is stationary

### KPSS Test

```python
from statsmodels.tsa.stattools import kpss
kpss_test = kpss(data['Number'])
print(kpss_test)
```

```python
print('KPSS statistic: %f' % kpss_test[0])
print('Critical Value @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %.2f' % kpss_test[1])
```

- p-value is less than 0.05, so the series is not stationary

### Box Cox Transformation

```python
from scipy.stats import boxcox
data_boxcox = pd.Series(boxcox(data['Number'], lmbda = 0), index=data.index)
```

```python
data_boxcox
```

```python
plt.figure(figsize=(12,4))
plt.plot(data_boxcox, label = 'After Box Cox Transformation')
plt.legend(loc='best')
plt.title('After Box Cox Transformation')
plt.show()
```

### Differencing

```python
data_boxcox_diff = pd.Series(data_boxcox - data_boxcox.shift(), index=data.index)
data_boxcox_diff
```

```python
data_boxcox_diff.dropna(inplace=True)
plt.figure(figsize=(12,4))
plt.plot(data_boxcox_diff, label='After box cox transform and differencing')
plt.legend(loc='best')
plt.title('After box cox transform and differencing')
plt.show()
```

- Now running ADF and KPSS test on this transformed series

```python
adf_test_boxcox = adfuller(data_boxcox_diff)
print(adf_test_boxcox)
```

```python
print('ADF statistic: %f' % adf_test_boxcox[0])
print('Critical Value @ 0.05: %.2f' % adf_test_boxcox[4]['5%'])
print('p-value: %f' % adf_test_boxcox[1])
```

```python
kpss_test_boxcox = kpss(data_boxcox_diff)
print(kpss_test_boxcox)
```

```python
print('KPSS statistic: %f' % kpss_test_boxcox[0])
print('Critical Value @ 0.05: %.2f' % kpss_test_boxcox[3]['5%'])
print('p-value: %.2f' % kpss_test_boxcox[1])
```

- Both the tests now verify that the series is stationary
